```python
import unidecode

def remove_accented_characters(text):
    """
    Remove accented characters from the given text.
    
    Parameters:
        text (str): Input text containing accented characters.
                    
    Returns:
        str: Text with accented characters removed.
        
    Example:
        Input : "Málaga, àéêöhello"
        Output : "Malaga, aeeohello"    
    """
    # Convert accented characters to their closest ASCII representation.
    return unidecode.unidecode(text)
     
```

```python
from autocorrect import Speller
```

```python
def spelling_correction(text):
    ''' 
    This function will correct spellings.
    
    Parameters:
         text (str): Input text containing misspelled words.
         
    Returns:
        str: Text after corrected spellings.
        
    Example: 
        Input : "This is Oberois from Dlhi who came heree to studdy."
        Output : "This is Oberoi from Delhi who came here to study."
    '''
    # Check for spellings in English language
    spell = Speller(lang='en')
    corrected_text = spell(text)
    return corrected_text
```

```python

def clean_ascii(text):
        # function to remove non-ASCII chars from data
        return ''.join(i for i in text if ord(i) < 128)
```

```python
import re
def remove_numbers(text):
    # define the pattern to keep
    pattern = r'[^a-zA-z.,!?:;\"\'\s]' 
    return re.sub(pattern, '', text)
```

```python
import contractions
def expand_contractions(text):
  return contractions.fix(text)
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
ps = PorterStemmer()
def preprocessing(text):
    text = remove_accented_characters(text)
    text = spelling_correction(text)
    text = clean_ascii(text)
    text = remove_numbers(text)
    text = expand_contractions(text)
    words = word_tokenize(text)
    stemmed_words = [ps.stem(word) for word in words]
    stemmed_sentence = ' '.join(stemmed_words)
    #print(stemmed_sentence[0])
    return stemmed_sentence
```

```python
random_rows = df.sample(n=5, replace=False, random_state=42)
random_rows['cleaned_articles'] = random_rows['article'].apply(preprocessing)
random_rows.head()
```
